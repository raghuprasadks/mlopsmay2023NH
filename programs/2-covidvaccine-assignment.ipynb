```python
import pandas as pd
print("1- covid dataframe")
df = pd.read_csv("covid_vaccine_statewise.csv")
```

```python
print("2-shape")
df.shape
```

```python
print('3. Display 20 % data')
df.sample(frac=.2)
```

```python
print('4. Display records for Karnataka State')
df.loc[df.State == "Karnataka"]
```

```python
print('5. Display total Covaxin doses administered for karnataka')
df.loc[df.State=="Karnataka"," Covaxin (Doses Administered)"].sum()
```

```python
print('6. Display total number doses for males and females administered for karnakata')
df.loc[df.State=="Karnataka","Male (Doses Administered)"].sum()
df.loc[df.State=="Karnataka","Female (Doses Administered)"].sum()
```

```python
print('7. Display total number of males and females administered dosage group by state')
df.groupby("State")["Male (Doses Administered)","Female (Doses Administered)"].agg(sum)
```
